```python
mbta = pd.read_excel('mbta.xlsx',header=1,index_col=0)
```

```python
mbta.head()
```

```python
mbta.info()
```

```python
#show all Nan Data in the first row
mbta.isna()
```

```python
#drop the first row
mbta.drop(index=[1], inplace = True)
```

```python
#Now no nan value
mbta.isna().sum()
```

```python
#Rows are best expressed as columns
mbta = mbta.melt(id_vars=['mode'], var_name='Date')
```

```python
mbta = mbta.pivot_table(index='Date',columns='mode', values='value')
```

```python
mbta
```

```python
#show data complex value , split it to two value month , year
#But Data must be col not id
```

```python
mbta.columns.name=None
mbta.reset_index(inplace = True)
```

```python

mbta
```

```python
#split Date to two cpl
mbta[['Year' , 'Month']]=mbta.Date.str.split('-', expand = True)
```

```python

mbta
```

```python
#now we don't need to Date now
mbta.drop('Date' , axis=1 , inplace =True)
```

```python
#we need to return the Year and Month to frist two row
mbta
```

```python
mbta = mbta[['Year',
         'Month',
         'Boat',
         'Bus',
         'Commuter Rail',
         'Heavy Rail',
         'Light Rail',
         'Private Bus',
         'RIDE',
         'Trackless Trolley']]
```

```python
#Now Data are Cleaning
mbta.head()
```

```python
#check if there outlier value
mbta['Boat'].describe()
```

```python
#show 40 are outlier value
mbta['Boat'].value_counts()
```

```python
#replace outlier value with Nan
mbta['Boat'].replace([40],np.nan,inplace=True)
```
